```python
units = {
    0: 'zero',
    1: 'one',
    2: 'two',
    3: 'three',
    4: 'four',
    5: 'five',
    6: 'six',
    7: 'seven',
    8: 'eight',
    9: 'nine',
    10: 'ten',
    11: 'eleven',
    12: 'twelve',
    13: 'thirteen',
    14: 'fourteen',
    15: 'fifteen',
    16: 'sixteen',
    17: 'seventeen',
    18: 'eighteen',
    19: 'nineteen'
}

dozens = {
    2: 'twenty',
    3: 'thirty',
    4: 'forty',
    5: 'fifty',
    6: 'sixty',
    7: 'seventy',
    8: 'eighty',
    9: 'ninety'
}
```

```python
for i in range(1000):
    num = str(i)
    if len(num) == 1:
        print(f'{i}: ', units[i])
    elif len(num) == 2:
        if i // 10 == 1:
            print(f'{i}: ', units[i])
        else:
            if i % 10 == 0:
                print(f'{i}: ', dozens[i // 10])
            else:
                print(f'{i}: ', f'{dozens[i // 10]} {units[i % 10]}')
    elif len(num) == 3:
        if i % 100 == 0:
            print(f'{i}: {units[i // 100]} hundred')
        elif i % 100 in units:
            print(f'{i}: {units[i // 100]} hundred and {units[i % 100]}')
        elif i % 10 == 0:
            print(f'{i}: {units[i // 100]} hundred and {dozens[(i // 10) % 10]}')
        else:
            print(f'{i}: {units[i // 100]} hundred and {dozens[(i // 10) % 10]} {units[i % 10]}')
```
