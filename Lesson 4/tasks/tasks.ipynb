# 1. БД пользователей

1. Нужно создать базу, где будет хранится информация о пользователях. В качестве БД будет выступать словарь (dict). Необходимаю информация о пользователе: `name`, `surname`, `age`, `sex`.
2. При запуске скрипта программа должна спросить, что будет происходить дальше:
    - поиск уже существещего пользователя;
    - запись нового пользователя.
3. Реализуйте поиск пользователя (используйте генераторы списков для поиска и форматирование строк для вывода информации):
    - по имени
    - по фамилии
    - по возрасту
    - по имени и фамилии
4. Реализуйте сохранение нового пользователя.
