```python
users = [
    {"name": 'Kseniya', "surname": 'Vorontsova', "age": 23, "sex": "female"},
    {"name": 'Vasya', "surname": 'Pupkin', "age": 33, "sex": "male"},
    {"name": 'Karinka', "surname": 'Petrovna', "age": 19, "sex": "female"}
]


def new_user():
    """
    Функция сохранения нового пользователя в базе
    :return: словарь с информацией о новом пользователе
    """

    print("Введите имя: ")
    name = input()

    print("Введите фамилию: ")
    surname = input()

    print("Введите возраст: ")
    age = input()

    print("Введите пол: ")
    sex = input()

    return {"name": name, "surname": surname, "age": age, "sex": sex}


def search(parameter):
    """
    Функция поиска по значению и полю
    :param parameter: поле поиска
    """
    found = []
    count_of_found = 0

    print("Введите значение")
    value = input()

    for item in users:
        if parameter == "age":
            if item[parameter] == int(value):# если в value ВВЕСТИ 18 и эа цифра есть  в item[parametr], то выводит что найден 1 ползователь
                found.append(item)
                count_of_found += 1
        else:
            if item[parameter] == value:
                found.append(item)
                count_of_found += 1

    print("Найдено пользователей: " + str(count_of_found))
    if count_of_found != 0:
        for found_element in found:
            string = f"""\
                    Name: {found_element["name"]} {found_element["surname"]}, 
                    age: {found_element["age"]}, sex: {found_element["sex"]}
                    """
            print(string)

print("------------")

# Бесконечный цикл, чтобы программа не закончилась без разрешения пользователя
ok = True

# пока ok = True, программа работает
while ok:
    print("Выберите функцию: ")
    print("1. Поиск")
    print("2. Новый пользователь")
    print("Завершить программу - любая другая кнопка")
    key = input()
    if int(key) == 1:
        print("Поиск пользователя:")
        print("N: по имени")
        print("S: по фамилии")
        print("A: по возрасту")
        print("G: по имени и фамилии")

        # p - параметр поиска, написано выше

        p = input()
        parameter = ""

        if p == 'N':
            parameter = "name"
            print("Поиск по имени")
            search(parameter=parameter)
        elif p == 'S':
            parameter = "surname"
            print("Поиск по фамилии")
            search(parameter=parameter)
        elif p == 'A':
            parameter = "age"
            print("Поиск по возрасту")
            search(parameter=parameter)
        elif p == 'G':
            print("Поиск по имени и фамилии")
            name_and_surname = input()

            # Разделение строки на имя и фамилию по пробелу - возвращает список ["Name", "Surname"]
            split_str = name_and_surname.split(" ")

            # Вывод найденных пользователей
            for item in users:
                if item["name"] == split_str[0] and item["surname"] == split_str[1]:
                    string = f"""\
                                        Name: {item["name"]} {item["surname"]}, 
                                        age: {item["age"]}, sex: {item["sex"]}
                                        """
                    print(string)
        else:
            print("Error")

    elif int(key) == 2:
        print("Добавить пользователя")

        # Добавляет new_user() в конец списка users
        users.append(new_user())
        print(users)
    else:
        print("Конец программы")
        ok = False
     
```
