# 1. Сеть

1. Написать фукцию, которая будет принимать IP адрес и маску подсети и вычислять IP адрес сети. IP адрес и маска должны быть в читаемом для человека виде, а все вычисления должны производится в 10-ой системе исчисления.
2. Написать декоратор для функции такой, что:

    - декорируемая функция принимает IP адрес и префикс маски в бинарном виде
    - декоратор проверяет на валидность IP адрес и префикс маски, если они не валидны, то возвращает 0
    - декоратор высчитывает из префикса маски маску в двоичном виде
    - декоратор переводит IP адрес и маску в читаемый для человека вид

# 2. Кэш функция

Необходимо реализовать декоратор, который будет сохранять результат выполнений функций с определённым набором параметров. Декоратор должен принимать аргумент, который отвечает за максимальное количество сохраняемых результатов.
1. декоратор сохраняет результат выполнения функции в глобальный словарь (ключ - пара аргументов, которые передаются функции, значение - результат выполнения функции)
2. декоратор должен проверить, вызывалась ли функция с данными параметрами, если да, то результат заново не высчитывается, а берётся из глобального словаря
3. если кэш заполнен (длина словаря больше указанного параметра для декоратора) `{новые аргументы: результат выполнения функции}` самый "старый" результат (для этого можно создать ещё один глобальный словарь, который будет хранить параметры функции как ключи, а время первого выполнения функции с такими параметрами - значение словаря)

Используйте следующие функцию *filter* и модуль *time*.
