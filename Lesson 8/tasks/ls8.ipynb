```python
with open('text.txt', 'r') as file:
    text = file.read()
    
import re
import collections

regex = r"[a-zA-Z]+(?<!'\w)"

matches = re.findall(regex, text, re.MULTILINE)

count_of_same_words = collections.Counter()


def counter_second():
    for i in matches:
        count_of_same_words[i] += 1
    return print('end count')


def low():
    for i in matches:
        matches[matches.index(i)] = matches[matches.index(i)].lower()
    print('end low')
    return matches
```

```python
print('Num of words:', len(matches))
low()
counter_second()
count_of_same_words.most_common(100)
```
