```python
with open('text.txt', 'r') as file:
    text = file.read()
    
import re

regex = r"[a-zA-Z]+"

matches = re.findall(regex, text, re.MULTILINE)

words = {}


def low():
    for i in matches:
        matches[matches.index(i)] = matches[matches.index(i)].lower()
    print('end low')
    return matches
    
    
def counter():
    for word in matches:
        if not word in words:
            a = len([i for i in matches if i == word])
            words.update({word: a})
        else:
            continue
    print('end count')
    return
    
    
def sort_list():
    list_d = list(words.items())
    list_d.sort(key=lambda i: i[1])
    for i in range(-1, -100, -1):
        print(list_d[i][0], ":",list_d[i][1])
    return
```

```python
print('Num of words:', len(matches))
low()
counter()
sort_list()
```
